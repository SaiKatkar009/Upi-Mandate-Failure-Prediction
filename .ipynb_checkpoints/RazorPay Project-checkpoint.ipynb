```python
import numpy as np
import pandas as pd
import random

from datetime import timedelta, datetime

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_customers = 5000
num_transactions = 50000

upi_apps = ['Gpay','PhonePe','Paytm']
upi_weights = [0.4, 0.35, 0.25] #Realistic Market Shares

banks = ['SBI', 'HDFC', 'ICICI', 'Axis', 'Kotak', 'BOB']
bank_weights = [0.25, 0.2, 0.2, 0.15, 0.1, 0.1] #Simulated Shares         

subscription_types = ['OTT','EMI','SaaS','Utility']
subscription_weights = [0.3, 0.4, 0.2, 0.1]

start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)
```

```python
customer_data = []
for customer_id in range(1, num_customers + 1):
    customer = {
        'customer_id': f'C{customer_id:05d}',
        'subscription_type': random.choices(subscription_types, subscription_weights)[0],
        'upi_app': random.choices(upi_apps, upi_weights)[0],
        'bank_name': random.choices(banks, bank_weights)[0],
        'mandate_start_date': start_date + timedelta(days=random.randint(0, 365))
    }
    customer_data.append(customer)

customers_df = pd.DataFrame(customer_data)
customers_df
```

```python
# Simulate Mandate Debit Attempts
transactions = []

for txn_id in range(1, num_transactions + 1):
    # Randomly pick a customer
    customer = customers_df.sample(1).iloc[0]

    # Random debit attempt date within mandate window
    debit_attempt_date = customer['mandate_start_date'] + timedelta(days=random.randint(0, 365))
    if debit_attempt_date > end_date:
        debit_attempt_date = end_date

    # Simulate account balance
    account_balance = np.round(random.uniform(0, 10000), 2)

    # Simulate bank API latency in milliseconds
    bank_latency = random.randint(100, 800)

    # Simulate bank success rate
    bank_success_rate = np.round(random.uniform(0.80, 0.99), 2)

    # Simulate mandate amount
    mandate_amount = random.choice([199, 499, 999, 1499, 1999, 2999])

    # Failure logic 
    failure_probability = 0.0
    if account_balance < mandate_amount * random.uniform(0.8, 1.2):
        failure_probability += 0.4 + random.uniform(0, 0.2)
    if bank_latency > random.randint(500, 800):
        failure_probability += 0.2
    if bank_success_rate < random.uniform(0.82, 0.9):
        failure_probability += 0.2


    # Random decision based on failure probability
    mandate_status = 1 if random.random() < failure_probability else 0  # 1 = Failure, 0 = Success

   
# Retry attempts not tied to mandate_status 
    retry_attempts = random.choices([0, 1, 2, 3], weights=[0.5, 0.2, 0.2, 0.1])[0]

    # Store the transaction
    transaction = {
        'transaction_id': f'TXN{txn_id:06d}',
        'customer_id': customer['customer_id'],
        'subscription_type': customer['subscription_type'],
        'upi_app': customer['upi_app'],
        'bank_name': customer['bank_name'],
        'mandate_amount': mandate_amount,
        'debit_attempt_date': debit_attempt_date,
        'account_balance': account_balance,
        'bank_api_latency': bank_latency,
        'bank_success_rate': bank_success_rate,
        'retry_attempts': retry_attempts,
        'mandate_status': mandate_status
    }

    transactions.append(transaction)

# Convert to DataFrame
transactions_df = pd.DataFrame(transactions)

# Preview
transactions_df
```

```python
transactions_df = transactions_df.sort_values(by=['customer_id', 'debit_attempt_date'])

transactions_df['mandate_age_days'] = 0
transactions_df['previous_failures'] = 0

customer_failure_count = {}

for idx, row in transactions_df.iterrows():
    cust_id = row['customer_id']
    mandate_start_date = customers_df.loc[customers_df['customer_id'] == cust_id, 'mandate_start_date'].values[0]

    mandate_age = (row['debit_attempt_date'] - pd.to_datetime(mandate_start_date)).days
    transactions_df.at[idx, 'mandate_age_days'] = mandate_age

    if cust_id not in customer_failure_count:
        customer_failure_count[cust_id] = 0

    transactions_df.at[idx, 'previous_failures'] = customer_failure_count[cust_id]
    
    if row['mandate_status'] == 1:
        customer_failure_count[cust_id] += 1

transactions_df.head()
```

## Model Building -

```python
feature_cols = ['mandate_amount', 'account_balance', 'bank_api_latency', 'bank_success_rate', 'retry_attempts', 'mandate_age_days', 'previous_failures']

X = transactions_df[feature_cols]
y = transactions_df['mandate_status']
```

```python
#Data Splitting

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20, random_state=42, stratify=y)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

logreg = LogisticRegression(max_iter= 1000)
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", conf_matrix)
```

```python
# Quick visual check
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(x='account_balance', y='mandate_amount', hue='mandate_status', data=transactions_df)
plt.title('Visual Check: Balance vs Mandate Amount by Mandate Status')
plt.show()
```

```python
transactions_df.groupby('mandate_status')['account_balance'].describe()
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(use_label_encoder= False, eval_metric='logloss')
xgb_model.fit(X_train, y_train)

y_pred_xgb = xgb_model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print("XGBoost Classification Report:\n")
print(classification_report(y_test, y_pred_xgb))

conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
print("\nConfusion Matrix:\n",conf_matrix_xgb)
```

```python
plt.figure(figsize=(10, 6))
xgb_importance = xgb_model.feature_importances_
features = X.columns

sns.barplot(x=xgb_importance, y=features, palette='viridis')
plt.title('Feature Importance from XGBoost Model')
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest Classification Report:\n")
print(classification_report(y_test, y_pred_rf))

conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix:\n", conf_matrix_rf)
```
